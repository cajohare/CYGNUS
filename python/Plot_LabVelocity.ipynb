```python
from numpy import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import LabFuncs

Location = array([50.0, 0.0])

nt = 1000
days = linspace(0.0,365.0,nt)
vlab_all = zeros(shape=(nt,3))
for i in range(0,nt):
    vlab_all[i,:] = LabFuncs.LabVelocity(days[i], Location)

    
    
fig = plt.figure(figsize=(8,7))
ax = fig.add_subplot(111)

plt.plot(days,vlab_all[:,0],'r-')
plt.plot(days,vlab_all[:,1],'b-')
plt.plot(days,vlab_all[:,2],'g-')


plt.show()
```
