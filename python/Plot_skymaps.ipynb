```python
from numpy import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import healpy as hp
plt.rcParams['axes.linewidth'] = 2.5
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

cmap = cm.Spectral_r
cmap.set_under('w')


RD_wimp0 = loadtxt('../code/RD_wimp_0deg.dat')
hp.mollview(sum(RD_wimp0,1)/amax(sum(RD_wimp0,1)),title="",cmap=cmap)
hp.graticule()

f = plt.gcf().get_children()
cbar = f[2]
cbar.set_xticklabels(['0', r'max(d$R$/d$E$d$\Omega$)'])  # vertically oriented colorbar
cbar.tick_params(labelsize=20) 
cbar.tick_params(which='major',direction='out',width=2,length=10,right=True,top=False,pad=7)
cbar.tick_params(which='minor',direction='out',width=1,length=7,right=True,top=False)

plt.savefig('../plots/skymap-0deg.pdf',bbox_inches='tight')
plt.show()
```
