```python
from numpy import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import healpy as hp
plt.rcParams['axes.linewidth'] = 2.5
plt.rc('text', usetex=True)
plt.rc('font', family='serif',size=20)

cmap = cm.Spectral_r
cmap.set_under('w')

RD_wimp0 = loadtxt('../code/RD_wimp_0deg.dat')
hp.mollview(flipud(sum(RD_wimp0,1)/amax(sum(RD_wimp0,1))),title=r"{\bf 10 GeV WIMP}",cmap=cmap,min=0,flip="geo")
hp.graticule(dpar=30.0,dmer=90.0,linewidth=3)
plt.text(0.01,0.02,r'$\hat{\mathcal{N}}$',fontsize=22)
plt.text(1,0.02,r'$\hat{\mathcal{W}}$',fontsize=22)
plt.text(-0.99,0.02,r'$\hat{\mathcal{E}}$',fontsize=22)
plt.text(-2.15,-0.01,r'$\hat{\mathcal{S}}$',fontsize=22)
plt.text(2.05,-0.01,r'$\hat{\mathcal{S}}$',fontsize=22)

f = plt.gcf().get_children()
cbar = f[2]
cbar.set_xticklabels(['0', r'max($\mathrm{d}^2R_\chi$/d$E$d$\Omega$)'])  # vertically oriented colorbar
cbar.tick_params(labelsize=20) 
cbar.tick_params(which='major',direction='out',width=2,length=10,right=True,top=False,pad=7)
cbar.tick_params(which='minor',direction='out',width=1,length=7,right=True,top=False)

plt.savefig('../plots/skymap-WIMP-0deg.pdf',bbox_inches='tight')
plt.show()

# Neutrinos
RD = loadtxt('../code/RD_nu_0deg.dat')
hp.mollview(flipud(sum(RD,1)/amax(sum(RD,1))),title=r"{\bf Solar neutrinos}",cmap=cmap,min=0,flip="geo")
hp.graticule(dpar=30.0,dmer=90.0)
plt.text(0.01,0.02,r'$\hat{\mathcal{N}}$',fontsize=22)
plt.text(1,0.02,r'$\hat{\mathcal{W}}$',fontsize=22)
plt.text(-0.99,0.02,r'$\hat{\mathcal{E}}$',fontsize=22)
plt.text(-2.15,-0.01,r'$\hat{\mathcal{S}}$',fontsize=22)
plt.text(2.05,-0.01,r'$\hat{\mathcal{S}}$',fontsize=22)

f = plt.gcf().get_children()
cbar = f[2]
cbar.set_xticklabels(['0', r'max($\mathrm{d}^2R_\nu$/d$E$d$\Omega$)'])  # vertically oriented colorbar
cbar.tick_params(labelsize=20) 
cbar.tick_params(which='major',direction='out',width=2,length=10,right=True,top=False,pad=7)
cbar.tick_params(which='minor',direction='out',width=1,length=7,right=True,top=False)

plt.savefig('../plots/skymap-Nu-0deg.pdf',bbox_inches='tight')
plt.show()
```
