```python
from numpy import *
from PlotFuncs import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import cmocean

RS = 1.0
Vs = '10k'
col1 = 'fuchsia'
col2 = 'blue'
eths1 = '_8keV'
eths2 = '_250eV'

def SIthplot1(fna,maxy1,maxy2,shx1=0,shy1=0,shx2=0,shy2=0):
    fig,ax = MakeLimitPlot_SI(Annotations=False,ymax=1.0e-38,ymin=1.0e-49,xmax=1.0e3,Collected=True)
    #plt.text(0.12,6.4e-46,r"$\nu$-{\bf floor}: F",fontsize=22,rotation=22,color="k")
    #plt.text(0.09,2e-45,r"$\nu$-{\bf floor}: He",fontsize=22,rotation=22,color="k")
    m,sig1 = getlimSI(Vs,fna+eths1,RS,maxy1)
    m,sig1w = get1wimpSI(Vs,fna+eths1,1.0,maxy1)
    plt.plot(m,sig1,linestyle='-',color='k',linewidth=5,zorder=11)
    plt.plot(m,sig1,linestyle='-',color=col1,linewidth=4,zorder=11)
    plt.plot(m,sig1w,linestyle='--',color=col1,linewidth=2,zorder=11)
    
    m,sig1 = getlimSI(Vs,fna+eths2,RS,maxy2)
    m,sig1w = get1wimpSI(Vs,fna+eths2,1.0,maxy2)
    plt.plot(m,sig1,linestyle='-',color='k',linewidth=5,zorder=11)
    plt.plot(m,sig1,linestyle='-',color=col2,linewidth=4,zorder=11)
    plt.plot(m,sig1w,linestyle='--',color=col2,linewidth=2,zorder=11)
    
    plt.gcf().text(0.88,0.83,r'{\bf {\sc Cygnus}-10k} m$^3$ $\times$ {\bf  6 yrs}',fontsize=40,color='w',horizontalalignment='right')
    plt.gcf().text(0.88,0.78,r'{\bf '+fna+r'}',fontsize=35,color='w',horizontalalignment='right')

    plt.gcf().text(0.36+shx1,0.74+shy1,r'{\bf Electron}',fontsize=25,color=col1,horizontalalignment='right')
    plt.gcf().text(0.36+shx1,0.71+shy1,r'{\bf discr.}',fontsize=25,color=col1,horizontalalignment='right')
    plt.gcf().text(0.36+shx1,0.68-0.005+shy1,r'$8$ keV$_{\rm r}$',fontsize=25,color=col1,horizontalalignment='right')


    plt.gcf().text(0.15+shx2,0.72+shy2,r'{\bf Single}',fontsize=25,color=col2)
    plt.gcf().text(0.143+shx2,0.69+shy2,r'{\bf electron}',fontsize=25,color=col2)
    plt.gcf().text(0.14+shx2,0.66-0.005+shy2,r'$0.25$ keV$_{\rm r}$',fontsize=25,color=col2)

    plt.plot(-1,-1,'-',color='k',lw=4,label=r'$90\%$ CL')
    plt.plot(-1,-1,'--',color='k',lw=2,label=r'$N_{\rm wimp} = 1$')

    leg1 = ax.legend(fontsize=30,frameon=True,loc="lower left",framealpha=1.0,ncol=1,columnspacing=0.5,labelspacing=0.1)
    leg1.get_frame().set_linewidth(3)
    leg1.get_frame().set_edgecolor("k")
    
    plt.show()

    fig.savefig('../plots/CYGNUS10k-'+fna+'_th_SI.pdf',bbox_inches='tight')
    return


# Enhancement factor
A = 19
S_p = 0.478
J = 0.5
C_F = (4/3)*((J+1)/J)*(S_p)**2.0/A**2

def SDthplot1(fna,maxy1,maxy2,shx1=0,shy1=0,shx2=0,shy2=0):
    fig,ax = MakeLimitPlot_SD(Annotations=False,xmax=1.0e3,Collected=True)
    plt.text(0.11,2e-38,r"$\nu$-{\bf floor}: Xe",fontsize=20,rotation=20,color="k")
    plt.text(0.13,0.3e-42,r"$\nu$-{\bf floor}: F",fontsize=20,rotation=20,color="k")
    
    m,sig1 = getlimSD(Vs,fna+eths1,C_F*1.5,maxy1*C_F)
    m,sig1w = get1wimpSD(Vs,fna+eths1,C_F*1.0,maxy1*C_F)
    plt.plot(m,sig1,linestyle='-',color='k',linewidth=5,zorder=11)
    plt.plot(m,sig1,linestyle='-',color=col1,linewidth=4,zorder=11)
    plt.plot(m,sig1w,linestyle='--',color=col1,linewidth=2,zorder=11)
    
    m,sig1 = getlimSD(Vs,fna+eths2,C_F*1.5,maxy2*C_F)
    m,sig1w = get1wimpSD(Vs,fna+eths2,C_F*1.0,maxy2*C_F)
    plt.plot(m,sig1,linestyle='-',color='k',linewidth=5,zorder=11)
    plt.plot(m,sig1,linestyle='-',color=col2,linewidth=4,zorder=11)
    plt.plot(m,sig1w,linestyle='--',color=col2,linewidth=2,zorder=11)
    plt.gcf().text(0.88,0.83,r'{\bf {\sc Cygnus}-10k}$^3$ $\times$ {\bf  6 yrs}',fontsize=40,color='w',horizontalalignment='right')
    plt.gcf().text(0.88,0.78,r'{\bf '+fna+r'}',fontsize=35,color='w',horizontalalignment='right')

    plt.gcf().text(0.40+shx1,0.67+shy1,r'{\bf Electron}',fontsize=25,color=col1,horizontalalignment='right')
    plt.gcf().text(0.40+shx1,0.64+shy1,r'{\bf discr.}',fontsize=25,color=col1,horizontalalignment='right')
    plt.gcf().text(0.40+shx1,0.61+shy1-0.005,r'$8$ keV$_{\rm r}$',fontsize=25,color=col1,horizontalalignment='right')


    plt.gcf().text(0.17+shx2,0.63+shy2,r'{\bf Single}',fontsize=25,color=col2)
    plt.gcf().text(0.163+shx2,0.6+shy2,r'{\bf electron}',fontsize=25,color=col2)
    plt.gcf().text(0.16+shx2,0.57+shy2-0.005,r'$0.25$ keV$_{\rm r}$',fontsize=25,color=col2)
    
    plt.plot(-1,-1,'-',color='k',lw=4,label=r'$90\%$ CL')
    plt.plot(-1,-1,'--',color='k',lw=2,label=r'$N_{\rm wimp} = 1$')

    leg1 = ax.legend(fontsize=30,frameon=True,loc="lower left",framealpha=1.0,ncol=1,columnspacing=0.5,labelspacing=0.1)
    leg1.get_frame().set_linewidth(3)
    leg1.get_frame().set_edgecolor("k")
    
    plt.show()
    fig.savefig('../plots/CYGNUS10k-'+fna+'_th_SD.pdf',bbox_inches='tight')
    return



```

```python
SIthplot1('Predrift',1.0e-39,1.0e-39)
SDthplot1('Predrift',1.0e-35,1.0e-39)
```

```python
SIthplot1('Postdrift',1.0e-39,1.0e-39,shy2=0.02,shx1=0.01)
SDthplot1('Postdrift',1.0e-35,1.0e-36)
```

```python
SIthplot1('Strip',1.0e-39,1.0e-39,shy2=0.08,shy1=-0.1)
SDthplot1('Strip',1.0e-35,1.0e-36,shx1=-0.02)
```

```python
fna = 'Pixel'
ths = array([1,2,3,4,5,6,7,8])
nths = size(ths)
cols = cm.rainbow(linspace(0.3,1.0,nths))

fig,ax = MakeLimitPlot_SI(Annotations=False,ymax=1.0e-38,ymin=1.0e-49,xmax=1.0e3,Collected=True)

maxy1 = 1.0e-30
for i in range(0,nths):
    eths = '_'+str(ths[i])+'keV'
    m,sig1 = getlimSI(Vs,fna+eths,RS,maxy1)
    m,sig1w = get1wimpSI(Vs,fna+eths,1.0,maxy1)
    ax.plot(m,sig1,linestyle='-',color='k',linewidth=5,zorder=11)
    ax.plot(m,sig1,linestyle='-',color=cols[i,:],linewidth=4,zorder=11,label=str(ths[i])+' keV')
    ax.plot(m,sig1w,linestyle='--',color=cols[i,:],linewidth=2,zorder=11)

plt.gcf().text(0.88,0.83,r'{\bf {\sc Cygnus}-10k} m$^3$ $\times$ {\bf  6 yrs}',fontsize=40,color='w',horizontalalignment='right')
plt.gcf().text(0.88,0.78,r'{\bf '+fna+r'}',fontsize=35,color='w',horizontalalignment='right')

#plt.gcf().text(0.36,0.74,r'{\bf Electron}',fontsize=25,color=col1,horizontalalignment='right')
#plt.gcf().text(0.36,0.71,r'{\bf discr.}',fontsize=25,color=col1,horizontalalignment='right')

#plt.plot(-1,-1,'-',color='k',lw=4,label=r'$90\%$ CL')
#plt.plot(-1,-1,'--',color='k',lw=2,label=r'$N_{\rm wimp} = 1$')
leg1 = ax.legend(fontsize=30,frameon=True,loc="lower left",framealpha=1.0,ncol=4,columnspacing=0.5,labelspacing=0.1)
leg1.get_frame().set_linewidth(3)
leg1.get_frame().set_edgecolor("k")

plt.show()

fig.savefig('../plots/CYGNUS10k-'+fna+'_th_SI.pdf',bbox_inches='tight')

```
