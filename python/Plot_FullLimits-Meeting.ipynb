# Plots CYGNUS for various readout params

```python
from numpy import *
from PlotFuncs import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import cmocean

fig,ax = MakeLimitPlot_SI(Annotations=False)
plt.text(0.14,6.4e-46,r"$\nu$-{\bf floor}: F",fontsize=22,rotation=22,color="k")
plt.text(0.11,2e-45,r"$\nu$-{\bf floor}: He",fontsize=22,rotation=22,color="k")

alph = 0.5
RS = 1.5
maxy = 1.0e-39
Vs = ['10','1000','100k']
cols = cm.Purples(linspace(0.7,1.0,size(Vs)))
for i in range(0,size(Vs)):
    m,sig1 = getlimSI(Vs[i],"Nondirectional_Final",RS,maxy)
    m,sig2 = getlimSI(Vs[i],"predrift_Final",RS,maxy)
    m,sig3 = getlimSI(Vs[i],"SearchMode-predrift",RS,maxy)
    
    sig3[sig3>sig2] = sig2[sig3>sig2]
    sig3[isnan(sig3)] = sig2[isnan(sig3)]
    
    plt.fill_between(m,sig3/RS,y2=sig1/RS,facecolor=cols[i,:],alpha=alph,zorder=10)
    plt.plot(m,sig2/RS,linestyle='-',color='k',linewidth=4,zorder=11)
    plt.plot(m,sig1/RS,linestyle='--',color=cols[i,:],linewidth=3,zorder=11)
    plt.plot(m,sig2/RS,linestyle='-',color=cols[i,:],linewidth=3,zorder=11)
    plt.plot(m,sig3/RS,linestyle='--',color=cols[i,:],linewidth=3,zorder=11)

plt.gcf().text(0.18,0.65,r'{\bf '+Vs[0]+r' m}$^3$',fontsize=28,color=cols[0,:],rotation=-40)
plt.gcf().text(0.15,0.59,r'{\bf '+Vs[1]+r' m}$^3$',fontsize=28,color=cols[1,:],rotation=-40)
plt.gcf().text(0.15,0.54,r'{\bf '+Vs[2]+r' m}$^3$',fontsize=28,color=cols[2,:],rotation=-40)
plt.gcf().text(0.88,0.83,r'{\bf {\sc Cygnus}-Nm}$^3$ $\times$ {\bf  6 yrs}',fontsize=40,color='w',horizontalalignment='right')


# hacked legend
plt.plot(m,-sig1,linestyle='--',color='w',linewidth=3,label='{\sc Cygnus} Non-directional sensitivity')
plt.plot(m,-sig2,linestyle='-',color='w',linewidth=3,label='{\sc Cygnus} Directional sensitivity')
plt.plot(m,-sig3,linestyle='--',color='w',linewidth=3,label='{\sc Cygnus} Search mode (50 Torr SF$_6$)')
leg1 = ax.legend(fontsize=30,frameon=True,loc="lower center",framealpha=1.0,ncol=1,columnspacing=0.5,labelspacing=0.1)
leg1.get_frame().set_linewidth(3)
leg1.get_frame().set_edgecolor("k")
x0 = 0.5
x1 = 1.29
yu = 3.12e-49
yl = 9.55e-51
sh = 0.3
plt.fill_between([x0,x1-sh/2.0],[yu,yu],y2=[yl,yl],color=cols[2,:],alpha=alph,linewidth=0,zorder=100)
plt.plot([x0,x1+sh],[yu,yu],'--',lw=3,zorder=100,color=cols[2,:])
plt.plot([x0,x1+sh],[yl,yl],'--',lw=3,zorder=100,color=cols[2,:])
plt.plot([x0+0.01,x1-0.03++sh],[10.0**(0.5*(log10(yl)+log10(yu))),10.0**(0.5*(log10(yl)+log10(yu)))],\
         '-',lw=3,zorder=100,color=cols[2,:])
plt.show()


fig.savefig('../plots/CYGNUS_V_SI.pdf',bbox_inches='tight')



```

```python
# Enhancement factor
A = 19
S_p = 0.478
J = 0.5
RS = (4/3)*((J+1)/J)*(S_p)**2.0/A**2

fig,ax = MakeLimitPlot_SD(Annotations=False)
plt.text(0.1,2e-38,r"$\nu$-{\bf floor}: Xe",fontsize=20,rotation=24,color="k")
plt.text(0.12,0.3e-42,r"$\nu$-{\bf floor}: F",fontsize=20,rotation=25,color="k")


alph = 0.5
maxy = 1.0e-35
Vs = ['10','1000','100k']
cols = cm.Purples(linspace(0.7,1.0,size(Vs)))
for i in range(0,size(Vs)):
    m,sig1 = getlimSD(Vs[i],"Nondirectional_Final",RS,maxy)
    m,sig2 = getlimSD(Vs[i],"predrift_Final",RS,maxy)
    m,sig3 = getlimSD(Vs[i],"SearchMode-predrift",RS,maxy)
    
    sig3[sig3>sig2] = sig2[sig3>sig2]
    sig3[isnan(sig3)] = sig2[isnan(sig3)]
    
    
    plt.fill_between(m,sig3,y2=sig1,facecolor=cols[i,:],alpha=alph,zorder=10)
    plt.plot(m,sig2,linestyle='-',color='k',linewidth=4,zorder=11)
    plt.plot(m,sig1,linestyle='--',color=cols[i,:],linewidth=3,zorder=11)
    plt.plot(m,sig2,linestyle='-',color=cols[i,:],linewidth=3,zorder=11)
    plt.plot(m,sig3,linestyle='--',color=cols[i,:],linewidth=3,zorder=11)

plt.gcf().text(0.23,0.7,r'{\bf '+Vs[0]+r' m}$^3$',fontsize=28,color=cols[0,:],rotation=-50)
plt.gcf().text(0.21,0.62,r'{\bf '+Vs[1]+r' m}$^3$',fontsize=28,color=cols[1,:],rotation=-50)
plt.gcf().text(0.21,0.54,r'{\bf '+Vs[2]+r' m}$^3$',fontsize=28,color=cols[2,:],rotation=-50)
plt.gcf().text(0.88,0.83,r'{\bf {\sc Cygnus}-Nm}$^3$ $\times$ {\bf  6 yrs}',fontsize=40,color='w',horizontalalignment='right')



# hacked legend
plt.plot(m,-sig1,linestyle='--',color='w',linewidth=3,label='{\sc Cygnus} Non-directional sensitivity')
plt.plot(m,-sig2,linestyle='-',color='w',linewidth=3,label='{\sc Cygnus} Directional sensitivity')
plt.plot(m,-sig3,linestyle='--',color='w',linewidth=3,label='{\sc Cygnus} Search mode (50 Torr SF$_6$)')
leg1 = ax.legend(fontsize=30,frameon=True,loc="lower center",framealpha=1.0,ncol=1,columnspacing=0.5,labelspacing=0.1)
leg1.get_frame().set_linewidth(3)
leg1.get_frame().set_edgecolor("k")
x0 = 0.5
x1 = 1.29
yu = 2.12e-46
yl = 9.55e-48
sh = 0.3
plt.fill_between([x0,x1-sh/2],[yu,yu],y2=[yl,yl],color=cols[2,:],alpha=alph,linewidth=0,zorder=100)
plt.plot([x0,x1+sh],[yu,yu],'--',lw=3,zorder=100,color=cols[2,:])
plt.plot([x0,x1+sh],[yl,yl],'--',lw=3,zorder=100,color=cols[2,:])
plt.plot([x0+0.01,x1-0.03++sh],[10.0**(0.5*(log10(yl)+log10(yu))),10.0**(0.5*(log10(yl)+log10(yu)))],'-',lw=3,zorder=100,color=cols[2,:])
plt.show()


plt.show()
fig.savefig('../plots/CYGNUS_V_SD.pdf',bbox_inches='tight')
```
