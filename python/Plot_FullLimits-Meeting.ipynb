```python
plt.rcParams['axes.linewidth'] = 2.5
plt.rc('text', usetex=True)
plt.rc('font', family='serif',size=25)

fig, ax1 = plt.subplots()
fig.set_size_inches(13,12)


xmin = 0.001
xmax = 1.0e10
ymax = 1.0e-41
ymin = 1.0e-51

C1 = loadtxt("../data/CYGNUS1000-Exposure-Ideal.txt")
NB8 = C1[3,:]
V = C1[0,:]
ax1.loglog(NB8,C1[1,:],'-',linewidth=3,color='k',label="Idealised")

C1 = loadtxt("../data/CYGNUS1000-Exposure-angres.txt")
ax1.loglog(NB8,C1[1,:],'.-',linewidth=3,color='navy',markersize=20,label="+Angular res.")

C1 = loadtxt("../data/CYGNUS1000-Exposure-angres-HT.txt")
ax1.loglog(NB8,C1[1,:],'.-',linewidth=3,color='crimson',markersize=20,label="+Angular res + head tail")

C1 = loadtxt("../data/CYGNUS1000-Exposure-Nondirectional.txt")
ax1.loglog(NB8,C1[1,:],'--',linewidth=3,color='k',label="Non-directional")

# Labels
ax1.set_xlabel(r"Number of neutrino events",fontsize=35)
ax1.set_ylabel(r"SI WIMP-proton cross section [cm$^2$]",fontsize=35)
ax1.set_xlim([NB8[0],NB8[-1]])

ax2 = ax1.twiny()  # instantiate a second axes that shares the same x-axis
ax2.set_xlabel(r'Volume [m$^3$]',fontsize=35,labelpad=15)  # we already handled the x-label with ax1
ax2.set_xlim([V[0],V[-1]])
ax2.loglog([1000,1000],[ymin,ymax],'k-')
ax2.loglog([1e5,1e5],[ymin,ymax],'k-')
ax2.loglog(-NB8,C1[1,:],'-',linewidth=3,color='k',label="Idealised pixel readout")
ax2.loglog(-NB8,C1[1,:],'.-',linewidth=3,color='navy',markersize=20,label="+Angular res.")
ax2.loglog(-NB8,C1[1,:],'.-',linewidth=3,color='crimson',markersize=20,label="+Angular res. + head-tail")
ax2.loglog(-NB8,C1[1,:],'--',linewidth=3,color='k',label="Non-directional")
ax2.text(400,3e-42,r'{\bf CYGNUS-1000}',fontsize=25,rotation=90)
ax2.text(40000,3e-42,r'{\bf CYGNUS-100k}',fontsize=25,rotation=90)

# Style
ax1.grid()
#ax2.grid()
ax1.set_yscale('log')
ax1.set_xscale('log')
ax2.set_xscale('log')
ax1.set_ylim(bottom=ymin, top=ymax)
ax1.tick_params(which='major',direction='in',width=2,length=13,right=True,top=False,pad=7)
ax1.tick_params(which='minor',direction='in',width=1,length=10,right=True,top=False)
ax2.tick_params(which='major',direction='in',width=2,length=13,bottom=False,top=True,pad=7)
ax2.tick_params(which='minor',direction='in',width=1,length=10,bottom=False,top=True)
leg1 = ax2.legend(fontsize=30,frameon=True,loc="lower left",framealpha=1)
leg1.get_frame().set_linewidth(3)
leg1.get_frame().set_edgecolor("k")

plt.show()
fig.savefig('../plots/CYGNUS1000-10GeVExposure.pdf',bbox_inches='tight')
```
